# Efficient Graph-Based Image Segmentation using Felzenszwalbâ€™s Algorithm

[Paper](http://people.cs.uchicago.edu/~pff/papers/seg-ijcv.pdf)

```python
import cv2
import numpy
import matplotlib.pyplot as plt
import numpy as np
import skimage.io
import skimage.feature
import skimage.color
import skimage.transform
import skimage.util
import skimage.segmentation
from skimage.segmentation import mark_boundaries
```

```python
img_path = "15297_front.jpg"
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
im_mask = skimage.segmentation.felzenszwalb(
        skimage.util.img_as_float(img), scale= 150, sigma= 1,
        min_size= 75)
bound_img = mark_boundaries(img, im_mask)
plt.figure(figsize=(20,20))
plt.subplot(231)
plt.imshow(img)
plt.subplot(232)
plt.imshow(im_mask)
plt.subplot(233)
plt.imshow(bound_img)
plt.show()
```
